```python
## Assuming df_renamed is your DataFrame

columns_to_process = [
    'ST',
    'RS',
    'LW',
    'LF',
    'CF',
    'RF',
    'RW',
    'LAM',
    'CAM',
    'RAM',
    'LM',
    'LCM',
    'CM',
    'RCM',
    'RM',
    'LWB',
    'LDM',
    'CDM',
    'RDM',
    'RWB',
    'LB',
    'LCB',
    'CB',
    'RCB',
    'RB'
]

# Loop through each column and apply the operation
for col in columns_to_process:
    df_cleaned[col] = df_cleaned[col].str.split('+').str[0]
#clearing the+ in the datasets
```

```python
df_view=df_cleaned[columns_to_process]
df_view
```

```python
new_column_names = {
    'LS': 'Left Striker',
    'ST': 'Striker',
    'RS': ' RightStriker',
    'LW': 'left Winger',
    'LF': 'left Forward',
    'CF': 'center Forward',
    'RF': 'right Forward',
    'RW': 'Right  Winger',
    'LAM': 'L Attacking Midfielder',
    'CAM': 'C Attacking Midfielder',
    'RAM': 'R Attacking Midfielder',
    'LM': 'Left Midfielder',
    'LCM': 'L Center Midfielder',
    'CM': 'Center Midfielder',
    'RCM': 'R Center Midfielder',
    'RM': 'Right Midfielder',
    'LWB': 'Left Wing Fullback',
    'LDM': 'L Defensive Midfielder',
    'CDM': 'C Defensive Midfielder',
    'RDM': 'R Defensive Midfielder',
    'RWB': 'Right Wing Fullback',
    'LB': 'Left Fullback',
    'LCB': 'L Center Back',
    'CB': 'Center Back or Sweeper',
    'RCB': 'R Center Back',
    'RB': 'Right Fullback'
}

# Rename the columns
df_renamed = df_cleaned.rename(columns=new_column_names)
```

```python
df_renamed
```

## opting out the euro symbol in the datasets

```python
df_renamed['Value'] = df_renamed['Value'].str.replace('€', '')
```

```python
df_renamed['Wage'] = df_renamed['Wage'].str.replace('€', '')
df_renamed['Release Clause'] = df_renamed['Release Clause'].str.replace('€', '')
```

```python
df_renamed['Left Striker'] = df_renamed['Left Striker'].str.split('+').str[0]
```

```python
df_renamed
```

```python
missing_values_renamed = df_renamed.isna().sum()

# Print the count of missing values for each column
print("Count of missing values for each column:")
print(missing_values)

# Total count of missing values
total_missing_renamed = missing_values_renamed.sum()
print("\nTotal missing values:", total_missing_renamed)
```

```python
# Using Forward fill NaN values for the "Release Clause" column
df_renamed['Release Clause'] = df_renamed['Release Clause'].fillna(method='ffill')
```

```python
#Saving the file as renamed_datset as csv file and not index so its false
df_renamed.to_csv('renamed_dataset.csv', index=False)
```
